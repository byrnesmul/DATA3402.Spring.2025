1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self, max_value):
        self.max_value = max_value
        self.value = 0
    
    def increment(self):
        if self.value < self.max_value:
            self.value += 1
        else:
            print("Error! Counter has reached its maximum.")
    
    def reset(self):
        self.value = 0
    
    def get_value(self):
        return self.value
    
# Testing the Counter class
c = Counter(3)
print(c.get_value())  # Expected output: 0

c.increment()
print(c.get_value())  # Expected output: 1

c.increment()
print(c.get_value())  # Expected output: 2

c.increment()
print(c.get_value())  # Expected output: 3

c.increment()  # Expected output: Error message
print(c.get_value())  # Expected output: 3

c.reset()
print(c.get_value())  # Expected output: 0
```

2. Copy and paste your solution to question 1 and modify it so that all the data held by the counter is private. Implement functions to check the value of the counter, check the maximum value, and check if the counter is at the maximum.

```python
class Counter:
    def __init__(self, max_value):
        self.__max_value = max_value
        self.__value = 0
    
    def increment(self):
        if self.__value < self.__max_value:
            self.__value += 1
        else:
            print("Error! Counter has reached its maximum.")
    
    def reset(self):
        self.__value = 0
    
    def get_value(self):
        return self.__value
    
    def get_max_value(self):
        return self.__max_value
    
    def is_at_max(self):
        return self.__value == self.__max_value
    
# Testing the Counter class
c = Counter(3)
print(c.get_value())  # Expected output: 0

c.increment()
print(c.get_value())  # Expected output: 1

c.increment()
print(c.get_value())  # Expected output: 2

c.increment()
print(c.get_value())  # Expected output: 3

c.increment()  # Expected output: Error message
print(c.get_value())  # Expected output: 3

print(c.get_max_value())  # Expected output: 3
print(c.is_at_max())  # Expected output: True

c.reset()
print(c.get_value())  # Expected output: 0
print(c.is_at_max())  # Expected output: False
```

6. Implement a triangle class analogous to the rectangle and circle in question 5.

```python
class Triangle(Shape):
    def __init__(self, side1, side2, side3, x, y):
        super().__init__(x, y)
        self.__side1 = side1
        self.__side2 = side2
        self.__side3 = side3
    
    def get_sides(self):
        return self.__side1, self.__side2, self.__side3
    
    def area(self):
        s = (self.__side1 + self.__side2 + self.__side3) / 2
        return math.sqrt(s * (s - self.__side1) * (s - self.__side2) * (s - self.__side3))
    
    def perimeter(self):
        return self.__side1 + self.__side2 + self.__side3
    
# Testing the Triangle class
t = Triangle(3, 4, 5, 0, 0)
print(t.get_sides())   
print(t.get_x())       
print(t.get_y())       
print(t.area())        # Expected output: 6.0
print(t.perimeter())   # Expected output: 12
```

```python
import math

class Shape:
    def __init__(self, x, y):
        self.__x = x
        self.__y = y
    
    def get_x(self):
        return self.__x
    
    def get_y(self):
        return self.__y
    
    def area(self):
        raise NotImplementedError("Subclasses must implement area method")
    
    def perimeter(self):
        raise NotImplementedError("Subclasses must implement perimeter method")
    
    def perimeter_points(self, num_points=16):
        raise NotImplementedError("Subclasses must implement perimeter_points method")

class Circle(Shape):
    def __init__(self, radius, x, y):
        super().__init__(x, y)
        self.__radius = radius
    
    def get_radius(self):
        return self.__radius
    
    def area(self):
        return math.pi * self.__radius ** 2
    
    def perimeter(self):
        return 2 * math.pi * self.__radius
    
    def perimeter_points(self, num_points=16):
        return [(self.get_x() + self.__radius * math.cos(2 * math.pi * i / num_points),
                 self.get_y() + self.__radius * math.sin(2 * math.pi * i / num_points))
                for i in range(num_points)]

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        super().__init__(x, y)
        self.__length = length
        self.__width = width
    
    def get_length(self):
        return self.__length
    
    def get_width(self):
        return self.__width
    
    def area(self):
        return self.__length * self.__width
    
    def perimeter(self):
        return 2 * (self.__length + self.__width)
    
    def perimeter_points(self, num_points=16):
        points = []
        edge_points = num_points // 4  
        
        # Bottom edge
        for i in range(edge_points):
            points.append((self.get_x() + i * (self.__length / edge_points), self.get_y()))
        
        # Right edge
        for i in range(edge_points):
            points.append((self.get_x() + self.__length, self.get_y() + i * (self.__width / edge_points)))
        
        # Top edge
        for i in range(edge_points):
            points.append((self.get_x() + self.__length - i * (self.__length / edge_points), self.get_y() + self.__width))
        
        # Left edge
        for i in range(edge_points):
            points.append((self.get_x(), self.get_y() + self.__width - i * (self.__width / edge_points)))
        
        return points

class Triangle(Shape):
    def __init__(self, x1, y1, x2, y2, x3, y3):
        super().__init__(x1, y1)  # Base corner
        self.__x2, self.__y2 = x2, y2
        self.__x3, self.__y3 = x3, y3
    
    def area(self):
        return abs((self.get_x() * (self.__y2 - self.__y3) +
                   self.__x2 * (self.__y3 - self.get_y()) +
                   self.__x3 * (self.get_y() - self.__y2)) / 2)
    
    def perimeter(self):
        return (math.dist((self.get_x(), self.get_y()), (self.__x2, self.__y2)) +
                math.dist((self.__x2, self.__y2), (self.__x3, self.__y3)) +
                math.dist((self.__x3, self.__y3), (self.get_x(), self.get_y())))
    
    def perimeter_points(self, num_points=16):
        points = []
        side_points = num_points // 3  # Distribute points among three sides
        
        def interpolate(x1, y1, x2, y2, count):
            return [(x1 + i * (x2 - x1) / count, y1 + i * (y2 - y1) / count) for i in range(count)]
        
        points += interpolate(self.get_x(), self.get_y(), self.__x2, self.__y2, side_points)
        points += interpolate(self.__x2, self.__y2, self.__x3, self.__y3, side_points)
        points += interpolate(self.__x3, self.__y3, self.get_x(), self.get_y(), side_points)
        
        return points

# Testing the improved perimeter_points functions
r = Rectangle(4, 3, 0, 0)
print("Rectangle Perimeter Points:", r.perimeter_points())

t = Triangle(0, 0, 4, 0, 2, 3)
print("Triangle Perimeter Points:", t.perimeter_points())

c = Circle(5, 0, 0)
print("Circle Perimeter Points:", c.perimeter_points())
```

8. Add a function to the object classes, including the base, that tests if a given set of $x$ and $y$ coordinates are inside of the object. You'll have to think through how to determine if a set of coordinates are inside an object for each object type.

```python
import math

class Shape:
    def __init__(self, x, y):
        self.__x = x
        self.__y = y
    
    def get_x(self):
        return self.__x
    
    def get_y(self):
        return self.__y
    
    def area(self):
        raise NotImplementedError("Subclasses must implement area method")
    
    def perimeter(self):
        raise NotImplementedError("Subclasses must implement perimeter method")
    
    def perimeter_points(self, num_points=16):
        raise NotImplementedError("Subclasses must implement perimeter_points method")
    
    def is_inside(self, x, y):
        raise NotImplementedError("Subclasses must implement is_inside method")

class Circle(Shape):
    def __init__(self, radius, x, y):
        super().__init__(x, y)
        self.__radius = radius
    
    def get_radius(self):
        return self.__radius
    
    def area(self):
        return math.pi * self.__radius ** 2
    
    def perimeter(self):
        return 2 * math.pi * self.__radius
    
    def perimeter_points(self, num_points=16):
        return [(self.get_x() + self.__radius * math.cos(2 * math.pi * i / num_points),
                 self.get_y() + self.__radius * math.sin(2 * math.pi * i / num_points))
                for i in range(num_points)]
    
    def is_inside(self, x, y):
        return math.dist((self.get_x(), self.get_y()), (x, y)) <= self.__radius

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        super().__init__(x, y)
        self.__length = length
        self.__width = width
    
    def get_length(self):
        return self.__length
    
    def get_width(self):
        return self.__width
    
    def area(self):
        return self.__length * self.__width
    
    def perimeter(self):
        return 2 * (self.__length + self.__width)
    
    def perimeter_points(self, num_points=16):
        points = []
        edge_points = num_points // 4  # Divide points among four edges
        
        # Bottom edge
        for i in range(edge_points):
            points.append((self.get_x() + i * (self.__length / edge_points), self.get_y()))
        
        # Right edge
        for i in range(edge_points):
            points.append((self.get_x() + self.__length, self.get_y() + i * (self.__width / edge_points)))
        
        # Top edge
        for i in range(edge_points):
            points.append((self.get_x() + self.__length - i * (self.__length / edge_points), self.get_y() + self.__width))
        
        # Left edge
        for i in range(edge_points):
            points.append((self.get_x(), self.get_y() + self.__width - i * (self.__width / edge_points)))
        
        return points

    def is_inside(self, x, y):
        return (self.get_x() <= x <= self.get_x() + self.__length and
                self.get_y() <= y <= self.get_y() + self.__width)

class Triangle(Shape):
    def __init__(self, x1, y1, x2, y2, x3, y3):
        super().__init__(x1, y1)  # Base corner
        self.__x2, self.__y2 = x2, y2
        self.__x3, self.__y3 = x3, y3
    
    def area(self):
        return abs((self.get_x() * (self.__y2 - self.__y3) +
                   self.__x2 * (self.__y3 - self.get_y()) +
                   self.__x3 * (self.get_y() - self.__y2)) / 2)
    
    def perimeter(self):
        return (math.dist((self.get_x(), self.get_y()), (self.__x2, self.__y2)) +
                math.dist((self.__x2, self.__y2), (self.__x3, self.__y3)) +
                math.dist((self.__x3, self.__y3), (self.get_x(), self.get_y())))
    
    def perimeter_points(self, num_points=16):
        points = []
        side_points = num_points // 3  # Distribute points among three sides
        
        def interpolate(x1, y1, x2, y2, count):
            return [(x1 + i * (x2 - x1) / count, y1 + i * (y2 - y1) / count) for i in range(count)]
        
        points += interpolate(self.get_x(), self.get_y(), self.__x2, self.__y2, side_points)
        points += interpolate(self.__x2, self.__y2, self.__x3, self.__y3, side_points)
        points += interpolate(self.__x3, self.__y3, self.get_x(), self.get_y(), side_points)
        
        return points

    def is_inside(self, x, y):
        # Helper function to calculate area of a triangle given three points
        def triangle_area(x1, y1, x2, y2, x3, y3):
            return abs((x1 * (y2 - y3) + x2 * (y3 - y1) + x3 * (y1 - y2)) / 2)

        A = self.area()  # Original triangle area
        A1 = triangle_area(x, y, self.__x2, self.__y2, self.__x3, self.__y3)
        A2 = triangle_area(self.get_x(), self.get_y(), x, y, self.__x3, self.__y3)
        A3 = triangle_area(self.get_x(), self.get_y(), self.__x2, self.__y2, x, y)

        return abs(A - (A1 + A2 + A3)) < 1e-6  # Allow for floating-point precision issues

# Testing the `is_inside` function
c = Circle(5, 0, 0)
print("Circle:")
print(c.is_inside(0, 0))  # Expected: True
print(c.is_inside(5, 5))  # Expected: False

r = Rectangle(4, 3, 0, 0)
print("Rectangle:")
print(r.is_inside(2, 1))  # Expected: True
print(r.is_inside(5, 5))  # Expected: False

t = Triangle(0, 0, 4, 0, 2, 3)
print("Triangle:")
print(t.is_inside(2, 1))  # Expected: True
print(t.is_inside(5, 5))  # Expected: False
```

9. Add a function in the base class of the object classes that returns true/false testing that the object overlaps with another object.

```python
import math

class Shape:
    def __init__(self, x, y):
        self.__x = x
        self.__y = y
    
    def get_x(self):
        return self.__x
    
    def get_y(self):
        return self.__y
    
    def area(self):
        raise NotImplementedError("Subclasses must implement area method")
    
    def perimeter(self):
        raise NotImplementedError("Subclasses must implement perimeter method")
    
    def is_inside(self, x, y):
        raise NotImplementedError("Subclasses must implement is_inside method")

    def overlaps(self, other):
        raise NotImplementedError("Subclasses must implement overlaps method")

class Circle(Shape):
    def __init__(self, radius, x, y):
        super().__init__(x, y)
        self.__radius = radius
    
    def get_radius(self):
        return self.__radius
    
    def area(self):
        return math.pi * self.__radius ** 2
    
    def perimeter(self):
        return 2 * math.pi * self.__radius
    
    def is_inside(self, x, y):
        return math.dist((self.get_x(), self.get_y()), (x, y)) <= self.__radius

    def overlaps(self, other):
        if isinstance(other, Circle):
            distance = math.dist((self.get_x(), self.get_y()), (other.get_x(), other.get_y()))
            return distance <= (self.__radius + other.get_radius())
        return False  # Extend for other shapes later

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        super().__init__(x, y)
        self.__length = length
        self.__width = width
    
    def get_length(self):
        return self.__length
    
    def get_width(self):
        return self.__width
    
    def area(self):
        return self.__length * self.__width
    
    def is_inside(self, x, y):
        return (self.get_x() <= x <= self.get_x() + self.__length and
                self.get_y() <= y <= self.get_y() + self.__width)

    def overlaps(self, other):
        if isinstance(other, Rectangle):
            return not (self.get_x() + self.__length < other.get_x() or
                        other.get_x() + other.get_length() < self.get_x() or
                        self.get_y() + self.__width < other.get_y() or
                        other.get_y() + other.get_width() < self.get_y())
        return False  # Extend for other shapes later

class Triangle(Shape):
    def __init__(self, x1, y1, x2, y2, x3, y3):
        super().__init__(x1, y1)  # Base corner
        self.__x2, self.__y2 = x2, y2
        self.__x3, self.__y3 = x3, y3
    
    def is_inside(self, x, y):
        def triangle_area(x1, y1, x2, y2, x3, y3):
            return abs((x1 * (y2 - y3) + x2 * (y3 - y1) + x3 * (y1 - y2)) / 2)

        A = self.area()
        A1 = triangle_area(x, y, self.__x2, self.__y2, self.__x3, self.__y3)
        A2 = triangle_area(self.get_x(), self.get_y(), x, y, self.__x3, self.__y3)
        A3 = triangle_area(self.get_x(), self.get_y(), self.__x2, self.__y2, x, y)

        return abs(A - (A1 + A2 + A3)) < 1e-6

    def area(self):
        return abs((self.get_x() * (self.__y2 - self.__y3) +
                   self.__x2 * (self.__y3 - self.get_y()) +
                   self.__x3 * (self.get_y() - self.__y2)) / 2)
    
    def overlaps(self, other):
        if isinstance(other, Triangle):
            # Check if any vertex of one triangle is inside the other
            if (self.is_inside(other.get_x(), other.get_y()) or
                self.is_inside(other.__x2, other.__y2) or
                self.is_inside(other.__x3, other.__y3) or
                other.is_inside(self.get_x(), self.get_y()) or
                other.is_inside(self.__x2, self.__y2) or
                other.is_inside(self.__x3, self.__y3)):
                return True
        return False  # Extend for other shapes later

# Testing
c1 = Circle(5, 0, 0)
c2 = Circle(4, 4, 4)
r1 = Rectangle(4, 3, 0, 0)
r2 = Rectangle(2, 2, 3, 2)
t1 = Triangle(0, 0, 4, 0, 2, 3)
t2 = Triangle(2, 1, 6, 1, 4, 4)

print("Circle Overlap:", c1.overlaps(c2))  # True or False
print("Rectangle Overlap:", r1.overlaps(r2))  # True or False
print("Triangle Overlap:", t1.overlaps(t2))  # True or False
```

10. Copy the `Canvas` class from lecture to in a python file creating a `paint` module. Copy your classes from above into the module and implement paint functions. Implement a `CompoundShape` class. Create a simple drawing demonstrating that all of your classes are working.

```python
import math

class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.data = [[' '] * width for _ in range(height)]

    def set_pixel(self, row, col, char='*'):
        if 0 <= row < self.height and 0 <= col < self.width:
            self.data[row][col] = char

    def display(self):
        print("\n".join("".join(row) for row in self.data))

    def clear_canvas(self):
        self.data = [[' '] * self.width for _ in range(self.height)]
    
    def v_line(self, x, y, length, char='*'):
        for i in range(length):
            self.set_pixel(x + i, y, char)

    def h_line(self, x, y, length, char='*'):
        for i in range(length):
            self.set_pixel(x, y + i, char)
            
    def line(self, x1, y1, x2, y2, char='*'):
        dx = x2 - x1
        dy = y2 - y1
        steps = max(abs(dx), abs(dy))
        for i in range(steps + 1):
            x = x1 + i * dx // steps
            y = y1 + i * dy // steps
            self.set_pixel(y, x, char)


class Shape:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def draw(self, canvas):
        raise NotImplementedError("Subclasses must implement draw method")

class Circle(Shape):
    def __init__(self, radius, x, y):
        super().__init__(x, y)
        self.radius = radius

    def draw(self, canvas):
        for i in range(canvas.height):
            for j in range(canvas.width):
                if math.isclose(math.dist((self.x, self.y), (j, i)), self.radius, abs_tol=0.5):
                    canvas.set_pixel(i, j, 'O')

class Rectangle(Shape):
    def __init__(self, width, height, x, y):
        super().__init__(x, y)
        self.width = width
        self.height = height

    def draw(self, canvas):
        for i in range(self.height):
            for j in range(self.width):
                if i == 0 or i == self.height - 1 or j == 0 or j == self.width - 1:
                    canvas.set_pixel(self.y + i, self.x + j, '#')

class Triangle(Shape):
    def __init__(self, x1, y1, x2, y2, x3, y3):
        super().__init__(x1, y1)
        self.x2, self.y2 = x2, y2
        self.x3, self.y3 = x3, y3

    def draw(self, canvas):
        canvas.line(self.x, self.y, self.x2, self.y2, '^')
        canvas.line(self.x2, self.y2, self.x3, self.y3, '^')
        canvas.line(self.x3, self.y3, self.x, self.y, '^')

class CompoundShape(Shape):
    def __init__(self, shapes):
        self.shapes = shapes

    def draw(self, canvas):
        for shape in self.shapes:
            shape.draw(canvas)


# Testing the module
if __name__ == "__main__":
    canvas = Canvas(60, 15)  # Increased width to ensure spacing

    # Adjusted positions to fully separate shapes
    circle = Circle(5, 8, 7)         # Left side
    rectangle = Rectangle(10, 5, 25, 3)  # Center (moved right)
    triangle = Triangle(45, 1, 50, 8, 40, 8)  # Right side (moved slightly right)

    compound_shape = CompoundShape([circle, rectangle, triangle])
    
    compound_shape.draw(canvas)
    canvas.display()
```

11. Create a `RasterDrawing` class. Demonstrate that you can create a drawing made of several shapes, paint the drawing, modify the drawing, and paint it again. 

```python
import math

class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.data = [[' '] * width for _ in range(height)]

    def set_pixel(self, row, col, char='*'):
        if 0 <= row < self.height and 0 <= col < self.width:
            self.data[row][col] = char

    def display(self):
        print("\n".join("".join(row) for row in self.data))

    def clear_canvas(self):
        self.data = [[' '] * self.width for _ in range(self.height)]
    
    def v_line(self, x, y, length, char='*'):
        for i in range(length):
            self.set_pixel(x + i, y, char)

    def h_line(self, x, y, length, char='*'):
        for i in range(length):
            self.set_pixel(x, y + i, char)
            
    def line(self, x1, y1, x2, y2, char='*'):
        dx = x2 - x1
        dy = y2 - y1
        steps = max(abs(dx), abs(dy))
        for i in range(steps + 1):
            x = x1 + i * dx // steps
            y = y1 + i * dy // steps
            self.set_pixel(y, x, char)


class Shape:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def draw(self, canvas):
        raise NotImplementedError("Subclasses must implement draw method")

class Circle(Shape):
    def __init__(self, radius, x, y):
        super().__init__(x, y)
        self.radius = radius

    def draw(self, canvas):
        for i in range(canvas.height):
            for j in range(canvas.width):
                if math.isclose(math.dist((self.x, self.y), (j, i)), self.radius, abs_tol=0.5):
                    canvas.set_pixel(i, j, 'O')

class Rectangle(Shape):
    def __init__(self, width, height, x, y):
        super().__init__(x, y)
        self.width = width
        self.height = height

    def draw(self, canvas):
        for i in range(self.height):
            for j in range(self.width):
                if i == 0 or i == self.height - 1 or j == 0 or j == self.width - 1:
                    canvas.set_pixel(self.y + i, self.x + j, '#')

class Triangle(Shape):
    def __init__(self, x1, y1, x2, y2, x3, y3):
        super().__init__(x1, y1)
        self.x2, self.y2 = x2, y2
        self.x3, self.y3 = x3, y3

    def draw(self, canvas):
        canvas.line(self.x, self.y, self.x2, self.y2, '^')
        canvas.line(self.x2, self.y2, self.x3, self.y3, '^')
        canvas.line(self.x3, self.y3, self.x, self.y, '^')

class CompoundShape(Shape):
    def __init__(self, shapes):
        self.shapes = shapes

    def draw(self, canvas):
        for shape in self.shapes:
            shape.draw(canvas)


# New RasterDrawing class
class RasterDrawing:
    def __init__(self, width, height):
        self.canvas = Canvas(width, height)
        self.shapes = []

    def add_shape(self, shape):
        """Add a new shape to the drawing."""
        self.shapes.append(shape)

    def remove_shape(self, shape):
        """Remove a shape from the drawing."""
        if shape in self.shapes:
            self.shapes.remove(shape)
        
    def paint(self):
        """Clear the canvas and redraw all shapes."""
        self.canvas.clear_canvas()
        for shape in self.shapes:
            shape.draw(self.canvas)
        self.canvas.display()


# Testing RasterDrawing
if __name__ == "__main__":
    drawing = RasterDrawing(60, 15)

    # Create and add shapes
    circle = Circle(5, 8, 7)
    rectangle = Rectangle(10, 5, 25, 3)
    triangle = Triangle(45, 1, 50, 8, 40, 8)

    drawing.add_shape(circle)
    drawing.add_shape(rectangle)
    drawing.add_shape(triangle)

    print("Initial Drawing:")
    drawing.paint()

    # Modify the drawing: Remove rectangle and add a new triangle
    drawing.remove_shape(rectangle)
    new_triangle = Triangle(15, 2, 20, 6, 10, 6)
    drawing.add_shape(new_triangle)

    print("\nModified Drawing:")
    drawing.paint()
```

```python
import math

class Shape:
    def __init__(self, x, y):
        self.__x = x
        self.__y = y
    
    def get_x(self):
        return self.__x
    
    def get_y(self):
        return self.__y
    
    def area(self):
        raise NotImplementedError("Subclasses must implement area method")
    
    def perimeter(self):
        raise NotImplementedError("Subclasses must implement perimeter method")
    
    def perimeter_points(self, num_points=16):
        raise NotImplementedError("Subclasses must implement perimeter_points method")
    
    def contains(self, x, y):
        raise NotImplementedError("Subclasses must implement contains method")
    
    def __repr__(self):
        return f"Shape({self.__x}, {self.__y})"

class Circle(Shape):
    def __init__(self, radius, x, y):
        super().__init__(x, y)
        self.__radius = radius
    
    def get_radius(self):
        return self.__radius
    
    def area(self):
        return math.pi * self.__radius ** 2
    
    def perimeter(self):
        return 2 * math.pi * self.__radius
    
    def perimeter_points(self, num_points=16):
        return [(self.get_x() + self.__radius * math.cos(2 * math.pi * i / num_points),
                 self.get_y() + self.__radius * math.sin(2 * math.pi * i / num_points))
                for i in range(num_points)]
    
    def contains(self, x, y):
        return (x - self.get_x())**2 + (y - self.get_y())**2 <= self.__radius**2
    
    def __repr__(self):
        return f"Circle({self.__radius}, {self.get_x()}, {self.get_y()})"

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        super().__init__(x, y)
        self.__length = length
        self.__width = width
    
    def get_length(self):
        return self.__length
    
    def get_width(self):
        return self.__width
    
    def area(self):
        return self.__length * self.__width
    
    def perimeter(self):
        return 2 * (self.__length + self.__width)
    
    def perimeter_points(self, num_points=16):
        points = [(self.get_x(), self.get_y()),
                  (self.get_x() + self.__length, self.get_y()),
                  (self.get_x() + self.__length, self.get_y() + self.__width),
                  (self.get_x(), self.get_y() + self.__width)]
        return points * (num_points // 4)
    
    def contains(self, x, y):
        return self.get_x() <= x <= self.get_x() + self.__length and self.get_y() <= y <= self.get_y() + self.__width
    
    def __repr__(self):
        return f"Rectangle({self.__length}, {self.__width}, {self.get_x()}, {self.get_y()})"

class Triangle(Shape):
    def __init__(self, side1, side2, side3, x, y):
        super().__init__(x, y)
        self.__side1 = side1
        self.__side2 = side2
        self.__side3 = side3
    
    def get_sides(self):
        return self.__side1, self.__side2, self.__side3
    
    def area(self):
        s = (self.__side1 + self.__side2 + self.__side3) / 2
        return math.sqrt(s * (s - self.__side1) * (s - self.__side2) * (s - self.__side3))
    
    def perimeter(self):
        return self.__side1 + self.__side2 + self.__side3
    
    def perimeter_points(self, num_points=16):
        return [(self.get_x(), self.get_y())] * num_points
    
    def contains(self, x, y):
        return False  # Implementing point-in-triangle logic is complex
    
    def __repr__(self):
        return f"Triangle({self.__side1}, {self.__side2}, {self.__side3}, {self.get_x()}, {self.get_y()})"

class RasterDrawing:
    def __init__(self, canvas):
        self.canvas = canvas
        self.shapes = []
    
    def add_shape(self, shape):
        self.shapes.append(shape)
    
    def save(self, filename):
        with open(filename, "w") as f:
            for shape in self.shapes:
                f.write(repr(shape) + "\n")  # Save each shape in Python constructor form
    
    @staticmethod
    def load(filename, canvas):
        drawing = RasterDrawing(canvas)
        with open(filename, "r") as f:
            for line in f:
                drawing.add_shape(eval(line.strip()))  # Recreate objects from file
        return drawing

# Testing the RasterDrawing class
def test_raster_drawing():
    class MockCanvas:
        def clear_canvas(self):
            print("Canvas cleared")

        def display(self):
            print("Canvas displayed")
    
    canvas = MockCanvas()
    drawing = RasterDrawing(canvas)
    
    # Create shapes and add them to drawing
    drawing.add_shape(Circle(5, 10, 10))
    drawing.add_shape(Rectangle(4, 6, 1, 2))
    drawing.add_shape(Triangle(3, 4, 5, 0, 0))
    
    # Save shapes
    drawing.save("shapes.txt")
    
    # Load shapes back
    loaded_drawing = RasterDrawing.load("shapes.txt", canvas)
    
    # Print loaded shapes
    print("Original Shapes:")
    for shape in drawing.shapes:
        print(shape)
        
    # Print loaded shapes to verify correctness
    print("\nLoaded Shapes:")
    for shape in loaded_drawing.shapes:
        print(shape)

# Run the test function
test_raster_drawing()


```
